# Hypothetical Study of Winston Churchill Speech

A rudimentary approach to studying the contents of the speech using a simple count of keywords. Each of the keywords could then be given a weight based on perceived relative importance.

```python
import numpy as np
import string
```

```python
#Text from this speech from:
#https://winstonchurchill.org/resources/speeches/1940-the-finest-hour/we-shall-fight-on-the-beaches/

file = open('We_shall_fight.txt', 'r') 
print file.read() 
```

```python
with open('We_shall_fight.txt', 'r') as file:
    text = file.readlines()
wordlist=[]
for line in text:
    words=line.split()
    wordlist.append(words)
#print wordlist
```

```python
def countKeyWord(keyword,speech=wordlist,dictionary=None,verbose=True):
    flat = [word.translate(None, string.punctuation)  for line in speech for word in line]
    num=flat.count(keyword)
    if verbose:
        print "Number of times \'"+keyword+"\' appears in this speech: "+str(num)
    if dictionary!=None:
        dictionary[keyword]=num
    return num
    
```

```python
countKeyWord("British")
```

```python
keywordlist=("British", "French","German","Dunkirk","Nazi","Hitler",'Russia',"America","Japan")
```

```python
for el in keywordlist:
    countKeyWord(el)
```

```python
arxiv={}
allwords = [word.translate(None, string.punctuation)  for line in wordlist for word in line]
for el in allwords:
    countKeyWord(el,wordlist,dictionary=arxiv,verbose=False)
```

```python
for key, value in sorted(arxiv.iteritems(), key=lambda (k,v): (v,k),reverse=True):
    print "%s: %s" % (key, value)
```
